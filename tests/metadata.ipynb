# Test meta-data

```python
# get_data_set_list()
datasets_info = beaapi.get_data_set_list(beaKey)
dataset_names = list(datasets_info['DatasetName'])
datasets_info
```

```python
# get_parameter_list()
param_infos = {dataset_name: beaapi.get_parameter_list(beaKey, dataset_name) for dataset_name in dataset_names}
for dataset_name, param_info in param_infos.items():
    print(dataset_name)
    display(param_info)
```

```python
# Make string  version for get_data()
print("{", end='')
for dataset_name, param_info in param_infos.items():
    if dataset_name!='APIDatasetMetaData':
        print("'{0}': {1},".format(dataset_name.lower(), str(list(param_info['ParameterName']))).lower())
print("}")
```

```python
search_result = beaapi.search_metadata('', beaKey)
search_result['Datasetname'].unique() #array(['FixedAssets', 'NIPA', 'NIUnderlyingDetail'], dtype=object)
```

```python
# beaapi.get_parameter_values()
param_vals = {}
for i, (dataset_name, param_info) in enumerate(param_infos.items()):
    #display(param_info)
    params = list(param_info['ParameterName'])
    for param in params:
        if (dataset_name, param) not in param_vals:
            param_vals[(dataset_name, param)] = beaapi.get_parameter_values(beaKey, dataset_name, param)

#display(param_vals[('APIDatasetMetaData', 'Dataset')])
for key, value in param_vals.items():
    print(str(key) + ": " + str(list(value.columns)))
```
